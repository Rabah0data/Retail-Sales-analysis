```python
import pandas as pd
df=pd.read_csv('C:\\Users\\IZIK\\Desktop\\data_analysis_portfolio\\projects\\retail_sales_analysis\\data\\customer_shopping_data.csv')
df['total']=df['quantity']*df['price']
df['invoice_data']=pd.to_datetime(df['invoice_date'], format='%d/%m/%Y')
df['months']=df['invoice_data'].dt.strftime('%b')
df.columns
```

```python
monthly_sale=df.groupby('months')['total'].sum()
category_sale=df.groupby('category')['total'].sum()
gender_sale=df.groupby('gender')['total'].sum()
ages=[0,18,25,35,45,55,65,100]
df['age_group']=pd.cut(df['age'],bins=ages)
ages_sale=df.groupby('age_group')['total'].sum()
category_quantity=df.groupby('category')['quantity'].sum()
shoppingmall_sale=df.groupby('shopping_mall')['total'].sum()
payment_type_sale=df.groupby('payment_method')['total'].sum()
```

```python
import matplotlib.pyplot as plt   
print(payment_type_sale)
plt.bar(payment_type_sale.index,payment_type_sale.values)
plt.xlabel('payment method')
plt.ylabel('total purchases')
```

```python
print(shoppingmall_sale)
plt.bar(shoppingmall_sale.index,shoppingmall_sale.values)
plt.xticks(rotation=90)
plt.xlabel('shopping mall')
plt.ylabel("total purchases")
plt.title("revenue by shopping mall")
```

```python
print(category_quantity)
plt.bar(category_quantity.index, category_quantity.values)
plt.xticks(rotation=90)
plt.xlabel("product category")
plt.ylabel("total quantity")
plt.title("quantity sold by each category")
```

```python
print(category_sale)
plt.bar(category_sale.index, category_sale.values)  
plt.xticks(rotation=90)
plt.xlabel("product category")
plt.ylabel("total purchase")
plt.title("revenue by product category")
```

```python
print(ages_sale)
plt.bar(ages_sale.index.astype(str), ages_sale.values)
plt.xlabel("ages ranges")
plt.ylabel("total purchase")
plt.title("purchase power distribution by ages")
```

```python
print(monthly_sale)
plt.plot(monthly_sale.index, monthly_sale.values)
plt.xticks(rotation=90)
plt.title('Monthly sales')
plt.xlabel('Months')
plt.ylabel('Total Sales')
```

```python
print(gender_sale)
plt.bar(gender_sale.index,gender_sale.values)
plt.title('gender purchase power distribution')
plt.xlabel('gender')
plt.ylabel('total purchase')
```
